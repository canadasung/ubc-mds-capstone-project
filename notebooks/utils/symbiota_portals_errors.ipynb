In another section of code, the documentation mentions that some Symbiota portals return a 404 error when the species isn't found. I tested this below, and found that that is not the case and all symbiota portals return 200 even if the species is non existant/not found. Instead, incorrect implementations of the symbiota APIs are likely responsible for the 404 errors.

```python
import requests

_SYMBIOTA_PORTALS = {
    "symbiota_mycoportal": "https://mycoportal.org/portal",
    "symbiota_lichen": "https://lichenportal.org/portal",
    "symbiota_bryophyte": "https://bryophyteportal.org/portal",
    "symbiota_cch2": "https://cch2.org/portal",
    "symbiota_sernec": "https://sernecportal.org/portal",
    "symbiota_nansh": "https://nansh.org/portal",
    "symbiota_macroalgae": "https://macroalgae.org/portal",
    "symbiota_pterido": "https://pteridoportal.org/portal",
    "symbiota_neherbaria": "https://neherbaria.org/portal",
    "symbiota_midatlantic": "https://midatlanticherbaria.org/portal",
}


def get_portals_erroring_on_missing(timeout: int = 15):

    for key, base_url in _SYMBIOTA_PORTALS.items():
        code = requests.get(
            f"{base_url}/api/v2/taxonomy/search",
            params={
                "taxon": "Not Species",
                "type": "EXACT",
                "limit": 1,
                "offset": 0,
            },
            headers={"User-Agent": "Mozilla/5.0", "Accept": "application/json"},
        ).status_code

        print(f"{key} returned {code} for nonexistant species")

get_portals_erroring_on_missing()
```
