# Symbiota Portal Sweep

Runs the same checks as `run_symbiota_check.py` across every known Symbiota portal.


```python
import sys, os
sys.path.insert(0, os.path.abspath("../.."))

from scripts.APIs_pipe.symbiota import SymbiotaAPI
```

```python
_SYMBIOTA_PORTALS = [
    # (portal_key, base_url, test_species)
    ("mycoportal",     "https://mycoportal.org/portal",              "Amanita muscaria"),       # iconic, widely collected fungus
    ("lichen",         "https://lichenportal.org/portal",            "Cladonia cristatella"),   # British soldiers, very common N. American lichen
    ("bryophyte",      "https://bryophyteportal.org/portal",         "Polytrichum commune"),    # common haircap moss, global distribution
    ("cch2",           "https://cch2.org/portal",                    "Quercus agrifolia"),      # coast live oak, quintessential California plant
    ("sernec",         "https://sernecportal.org/portal",            "Magnolia grandiflora"),   # southern magnolia, iconic SE US tree
    ("nansh",          "https://nansh.org/portal",                   "Rudbeckia hirta"),        # black-eyed Susan, widespread N. American forb
    ("macroalgae",     "https://macroalgae.org/portal",              "Fucus vesiculosus"),      # bladderwrack, abundant N. Atlantic seaweed
    ("pterido",        "https://pteridoportal.org/portal",           "Pteridium aquilinum"),   # bracken fern, most widespread fern on Earth
    ("neherbaria",     "https://neherbaria.org/portal",              "Acer rubrum"),            # red maple, ubiquitous New England tree
    ("midatlantic",    "https://midatlanticherbaria.org/portal",     "Quercus rubra"),          # northern red oak, dominant mid-Atlantic hardwood
    ("swbiodiversity", "https://swbiodiversity.org/seinet",          "Larrea tridentata"),      # creosote bush, defining SW desert shrub
]
```

## Helper

```python
def check_portal(portal_url: str, species_name: str) -> None:
    api = SymbiotaAPI(portal_url)
    print(f"  portal_name : {api.portal_name}")

    print(f"\n  search('{species_name}')")
    result = api.search(species_name)
    if result is None:
        print("  status      : FAILED (all endpoints returned None)")
    else:
        results_list = result.get("results", [])
        print(f"  status      : OK — {len(results_list)} result(s)")
        print(f"  results     : {results_list}")

    print(f"\n  synonyms('{species_name}')")
    df = api.synonyms(species_name)
    print(f"  shape       : {df.shape}")
    print(df.to_string(index=False))
```

## Sweep all portals

```python
for name, url, species in _SYMBIOTA_PORTALS:
    print(f"\n{'='*60}")
    print(f"  PORTAL: {name}  ({url})")
    print('='*60)
    try:
        check_portal(url, species)
    except Exception as e:
        print(f"  ERROR: {e}")
```
