```python
from scripts.APIs_pipe.col import COLAPI
from scripts.APIs_pipe.gbif import GBIFAPI
from scripts.APIs_pipe.genbank import GenBankAPI
from scripts.APIs_pipe.index_fungorum import IndexFungorumAPI
from scripts.APIs_pipe.mushroomobs import MushroomObserverAPI
from scripts.APIs_pipe.symbiota import SymbiotaAPI
from scripts.APIs_pipe.tropicos import TropicosAPI
```

```python
# ── CONFIGURE HERE ──────────────────────────────────────────────────────────
SPECIES_MAP = {
    "gbif":                    "Amanita muscaria",       # general biodiversity
    "col":                     "Quercus robur",          # general taxonomy backbone
    "tropicos":                "Quercus robur",          # plant-focused taxonomy
    "index_fungorum":          "Amanita muscaria",       # fungi taxonomy
    "mushroomobs":             "Amanita muscaria",       # mushroom observations
    "genbank":                 "Amanita muscaria",       # genetic sequences (always empty)
    "symbiota_mycoportal":     "Amanita muscaria",       # fungi portal
    "symbiota_lichen":         "Cladonia rangiferina",   # lichen portal
    "symbiota_bryophyte":      "Sphagnum magellanicum",  # moss/bryophyte portal
    "symbiota_cch2":           "Quercus agrifolia",      # California plants (CCH2)
    "symbiota_sernec":         "Magnolia grandiflora",   # SE US herbaria
    "symbiota_nansh":          "Solidago canadensis",    # N American native plants
    "symbiota_macroalgae":     "Macrocystis pyrifera",   # macroalgae (giant kelp)
    "symbiota_pterido":        "Pteridium aquilinum",    # ferns & allies
    "symbiota_neherbaria":     "Acer saccharum",         # NE US herbaria
    "symbiota_midatlantic":    "Kalmia latifolia",       # Mid-Atlantic herbaria
    "symbiota_swbiodiversity": "Larrea tridentata",      # SW US / SEINet
}
# ────────────────────────────────────────────────────────────────────────────
```

```python
_SYMBIOTA_PORTALS = [
    ("mycoportal",    "https://mycoportal.org/portal"),
    ("lichen",        "https://lichenportal.org/portal"),
    ("bryophyte",     "https://bryophyteportal.org/portal"),
    ("cch2",          "https://cch2.org/portal"),
    ("sernec",        "https://sernecportal.org/portal"),
    ("nansh",         "https://nansh.org/portal"),
    ("macroalgae",    "https://macroalgae.org/portal"),
    ("pterido",       "https://pteridoportal.org/portal"),
    ("neherbaria",    "https://neherbaria.org/portal"),
    ("midatlantic",   "https://midatlanticherbaria.org/portal"),
    ("swbiodiversity","https://swbiodiversity.org/seinet"),
]

ALL_CLIENTS = {
    "gbif":           GBIFAPI(),
    "col":            COLAPI(),
    "tropicos":       TropicosAPI(),
    "index_fungorum": IndexFungorumAPI(),
    "mushroomobs":    MushroomObserverAPI(),
    "genbank":        GenBankAPI(),
    **{f"symbiota_{name}": SymbiotaAPI(url) for name, url in _SYMBIOTA_PORTALS},
}

print(f"Clients built: {list(ALL_CLIENTS)}")
```

```python
from IPython.display import display

for key, client in ALL_CLIENTS.items():
    species = SPECIES_MAP[key]
    print(f"\n{'='*60}")
    print(f"  {key}  ({client.__class__.__name__})")
    print(f"  Species: {species}")
    print(f"{'='*60}")
    try:
        result = client.synonyms(species)
        print(f"Rows returned: {len(result)}")
        display(result)
    except Exception as e:
        print(f"ERROR: {e}")
```
