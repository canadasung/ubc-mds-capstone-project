## Symbiota Portals

```python
_SYMBIOTA_PORTALS = [
    # (portal_key, base_url, species)
    ("mycoportal",     "https://mycoportal.org/portal",              "Agaricus campestris"),      # field mushroom, common, widespread
    ("lichen",         "https://lichenportal.org/portal",            "Xanthoria parietina"),  # common orange lichen, global distribution
    ("bryophyte",      "https://bryophyteportal.org/portal",         "Polytrichum commune"),   # common haircap moss, global distribution
    ("cch2",           "https://cch2.org/portal",                    "Quercus agrifolia"),     # coast live oak, quintessential California plant
    ("sernec",         "https://sernecportal.org/portal",            "Magnolia grandiflora"),  # southern magnolia, iconic SE US tree
    ("nansh",          "https://nansh.org/portal",                   "Rudbeckia hirta"),       # black-eyed Susan, widespread N. American forb
    ("macroalgae",     "https://macroalgae.org/portal",              "Fucus vesiculosus"),     # bladderwrack, abundant N. Atlantic seaweed
    ("pterido",        "https://pteridoportal.org/portal",           "Dryopteris filix-mas"),   # male fern, common, widespread fern
    ("neherbaria",     "https://neherbaria.org/portal",              "Acer rubrum"),           # red maple, ubiquitous New England tree
    ("midatlantic",    "https://midatlanticherbaria.org/portal",     "Quercus rubra"),         # northern red oak, dominant mid-Atlantic hardwood
    ("swbiodiversity", "https://swbiodiversity.org/seinet",          "Larrea tridentata"),     # creosote bush, defining SW desert shrub
]
```

```python
for portal_key, base_url, species in _SYMBIOTA_PORTALS:
    print(f"\n{'='*60}")
    print(f"  symbiota / {portal_key}")
    print(f"  URL:     {base_url}")
    print(f"  Species: {species}")
    print(f"{'='*60}")
    try:
        client = SymbiotaAPI(base_url=base_url, portal_name=portal_key)
        result = client.get_synonyms(species)
        print(f"Synonyms returned: {len(result)}")
        display(result)
    except Exception as e:
        print(f"ERROR: {e}")
```
