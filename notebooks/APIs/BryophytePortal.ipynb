# `get_bryophyteportal_synonyms` tests

`get_bryophyteportal_synonyms(species_name)` returns a dict whose keys are species-level synonym names and values are empty lists.

It uses three steps internally:
1. **gettaxasuggest RPC** — convert the species name to a taxon ID (tid)
2. **REST API** — resolve to the accepted taxon's tid if the input is itself a synonym
3. **HTML page** — scrape species-level synonyms from the portal's `synonymDiv`

```python
import sys, os
sys.path.insert(0, os.path.abspath("../.."))

from scripts.APIs.BryophytePortal import get_bryophyteportal_synonyms
```

## Species with multiple synonyms

External link to cross-check: [https://bryophyteportal.org/portal/taxa/index.php?taxon=Marchantia+polymorpha](https://bryophyteportal.org/portal/taxa/index.php?taxon=Marchantia+polymorpha)

```python
get_bryophyteportal_synonyms("Marchantia polymorpha")
```

## Species with no species-level synonyms

BryophytePortal has no synonyms listed for *Fissidens aphelotaxifolius* — the synonymDiv is absent, so only the queried name is returned.

External link to cross-check: [https://bryophyteportal.org/portal/taxa/index.php?taxon=Fissidens+aphelotaxifolius](https://bryophyteportal.org/portal/taxa/index.php?taxon=Fissidens+aphelotaxifolius)

```python
get_bryophyteportal_synonyms("Fissidens aphelotaxifolius")
```

## Input is itself a synonym

*Marchantia alpestris* is a synonym of *Marchantia polymorpha*. The function resolves to the accepted taxon first, then returns all synonyms of *Marchantia polymorpha*. Same synonym list as querying *Marchantia polymorpha* directly — except *Marchantia alpestris* itself is listed first (as the queried name) and any duplicate from the scraped page is deduplicated.

External link to cross-check: [https://bryophyteportal.org/portal/taxa/index.php?taxon=Marchantia+alpestris#](https://bryophyteportal.org/portal/taxa/index.php?taxon=Marchantia+alpestris#)

```python
get_bryophyteportal_synonyms("Marchantia alpestris")
```
