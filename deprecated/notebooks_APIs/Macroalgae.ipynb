# `get_macroalgae_synonyms` tests

`get_macroalgae_synonyms(species_name)` returns a dict whose keys are species-level synonym names and values are empty lists.

It uses three steps internally:
1. **gettaxasuggest RPC** — convert the species name to a taxon ID (tid)
2. **REST API** — resolve to the accepted taxon's tid if the input is itself a synonym
3. **HTML page** — scrape species-level synonyms from the portal's `synonymDiv`

```python
import sys, os
sys.path.insert(0, os.path.abspath("../.."))

from scripts.APIs.Macroalgae import get_macroalgae_synonyms
```

## Species with multiple synonyms

External link to cross-check: [https://macroalgae.org/portal/taxa/index.php?taxon=Ulva+lactuca](https://macroalgae.org/portal/taxa/index.php?taxon=Ulva+lactuca)

```python
get_macroalgae_synonyms("Ulva lactuca")
```

## Species with no species-level synonyms

The Macroalgae Portal has no synonyms listed for *Ulva ranunculata* — the synonymDiv is absent, so only the queried name is returned.

External link to cross-check: [https://macroalgae.org/portal/taxa/index.php?taxon=Ulva+ranunculata](https://macroalgae.org/portal/taxa/index.php?taxon=Ulva+ranunculata)

```python
get_macroalgae_synonyms("Ulva ranunculata")
```

## Input is itself a synonym

*Phyllona lactuca* is a synonym of *Ulva lactuca*. The function resolves to the accepted taxon first, then returns all synonyms of *Ulva lactuca*. Same synonym list as querying *Ulva lactuca* directly — except *Phyllona lactuca* is listed first (as the queried name) and any duplicate from the scraped page is deduplicated.

External link to cross-check: [https://macroalgae.org/portal/taxa/index.php?taxon=Phyllona+lactuca](https://macroalgae.org/portal/taxa/index.php?taxon=Phyllona+lactuca)

```python
get_macroalgae_synonyms("Phyllona lactuca")
```
