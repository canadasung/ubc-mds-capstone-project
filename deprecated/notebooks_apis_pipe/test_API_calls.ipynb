```python
from IPython.display import display

for key, client in ALL_CLIENTS.items():
    species = SPECIES_MAP[key]
    print(f"\n{'='*60}")
    print(f"  {key}  ({client.__class__.__name__})")
    print(f"  Species: {species}")
    print(f"{'='*60}")
    try:
        result = client.synonyms(species)
        print(f"Rows returned: {len(result)}")
        display(result)
    except Exception as e:
        print(f"ERROR: {e}")
```
