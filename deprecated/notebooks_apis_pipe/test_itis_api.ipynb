```python
STANDARD_QUERIES = [
    ("itis",           ITISAPI(),              "Rosa canina"), # Vascular plant
    ("itis",           ITISAPI(),              "Apis mellifera"), # Insect
    ("itis",           ITISAPI(),              "Macrocystis pyrifera"), # Algae
    ("itis",           ITISAPI(),              "Acer grandidentatum var. sinuosum"), # Algae
    ("itis",           ITISAPI(),              "Acer saccharum ssp. brachypterum"), # Algae
    ("itis",           ITISAPI(),              "Danaus plexippus"), # Algae
    ("itis",           ITISAPI(),              "Spirogyra porticalis"), # Algae
    ("itis",           ITISAPI(),              "Chondrus crispus"), # Algae
    ("itis",           ITISAPI(),              "Dunaliella salina"), # Algae
    ("itis",           ITISAPI(),              "Amanita muscaria"), # Fungi
]
```

```python
for key, client, species in STANDARD_QUERIES:
    print(f"\n{'='*60}")
    print(f"  {key}  ({client.__class__.__name__})")
    print(f"  Species: {species}")
    print(f"{'='*60}")
    try:
        result = client.get_synonyms(species)
        print(f"Synonyms returned: {len(result)}")
        display(result)
    except Exception as e:
        print(f"ERROR: {e}")
```
